```python
#import pacakges 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#display graphs in notebook
%matplotlib inline
sns.set()
#read excel workbook into dataframe

data = pd.read_excel('appl.xlsx')
```

```python
data.head()
```

```python
applClose = sns.lineplot(x="date", y="close", data=data,).set_title("Apple Closing Price")

fig = applClose.get_figure()
fig.set_size_inches(12, 8)
fig.savefig('AppleClosingPrice.png')
```

```python
mda50 = sns.lineplot(x="date", y="50 mda", data=data).set_title("Apple 50 day moving average")

fig = mda50.get_figure()
fig.set_size_inches(12, 8)
fig.savefig('mda50.png')
```

```python
techEtf = sns.lineplot(x="date", y="etf close", data=data)
```

```python
sandp500 = sns.lineplot(x="date", y="s&p 500 close", data=data)
```

```python
correlation = data.corr(method='pearson')
```
